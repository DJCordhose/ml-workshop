## Pie Chart
* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html
* https://matplotlib.org/examples/pie_and_polar_charts/pie_demo_features.html

```python
# Example data

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
```

```python
fig, ax = plt.subplots(4, figsize=(20,20))

ax[0].pie(x = sizes)

ax[1].pie(x = sizes, labels=labels, autopct='%1.1f%%')
ax[1].axis('equal')

ax[2].pie(x = sizes, autopct='%1.1f%%', explode=(0, 0.1, 0, 0), shadow=True, startangle=90)
ax[2].axis('equal')
ax[2].legend(labels=labels)

ax[3].pie(x = sizes, wedgeprops={'width': 0.5})
ax[3].axis('equal')

fig.show()
```

## Bar Chart
* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html
* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.barh.html

```python
# Example data

people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
performance2 = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
```

```python
fig, ax = plt.subplots(4, figsize=(20, 20))

ax[0].bar(x = pos, height = performance)

ax[1].barh(y = pos, width = performance, color='green')

ax[2].barh(y = pos, width = performance, xerr=error, color='orange')
ax[2].set_yticks(pos)
ax[2].set_yticklabels(people)
ax[2].set_xlabel('Person')
ax[2].set_ylabel('Performance')

bar_width = 0.35
ax[3].bar(x = pos, height = performance, width = bar_width, label='Performance 1')
ax[3].bar(x = pos + bar_width, height = performance2, width = bar_width, label='Performance 1')
ax[3].legend()
ax[3].set_title('Two Measurements')
ax[3].set_xticks(pos + bar_width / 2)
ax[3].set_xticklabels(people)

fig.savefig('bars.png')

fig.show()
```

## Histogram
* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html
* https://matplotlib.org/examples/statistics/histogram_demo_multihist.html

```python
# Example data

mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(500)
xs = np.random.randn(500, 3)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax0, ax1, ax2, ax3 = axes.flatten()

ax0.hist(x=x)

ax1.hist(x=x, bins=50)

ax2.hist(x=xs, color=['red', 'yellow', 'green'])

ax3.hist(x=xs, stacked=True, cumulative=True)

fig.show()
```
