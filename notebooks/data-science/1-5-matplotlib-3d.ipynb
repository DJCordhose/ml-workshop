```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# for interactive 3d on local jupyer notebook
%matplotlib notebook
```

```python
from sklearn import datasets
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_points = 1000
X, _ = datasets.samples_generator.make_s_curve(n_points, random_state=0)
```

```python
fig = plt.figure(figsize=(8, 8))
fig.suptitle("The disconnected S Curve with %i points"
             % (n_points), fontsize=20)

# ax = fig.add_subplot(111, projection='3d')
# ax = fig.add_subplot(1, 1, 1, projection='3d')
nrows=1
ncols=1
index=1
ax = fig.add_subplot(nrows, ncols, index, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2])

# https://en.wikipedia.org/wiki/Azimuth
ax.view_init(elev=10, azim=-60)
```
